#### Ridge regression

```python
def ridge(X, y, lam):
    clf = Ridge(alpha=lam)
    clf.fit(X.values, y)
    clf.predict(y)
    score = clf.score(X, y, sample_weight = None)
    nonzero = (clf.coef_ != 0)
    # weights = clf.coef_
    # cols = X.loc[:, z].columns.values.tolist()
    print("Number of non-zero features: " + str(len(nonzero)) + ", alpha: " + str(lam))
    return score
```

```python
err = ridge(X,y,1)
```
